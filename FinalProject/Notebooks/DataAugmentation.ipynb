```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import cv2
import os
```

```python
CAREGORIES = ['ChayLa']
DATA_DIR = '/content/drive/MyDrive/CS114.M21.KHCL/DoAn/Data/Train'

SAVE_DIR = '/content/drive/MyDrive/CS114.M21.KHCL/DoAn/Data/TrainAug'
```

```python
for CATEGORY in CAREGORIES:
    path = os.path.join(DATA_DIR, CATEGORY)
    j = 0
    for img_path in os.listdir(path):
        img = cv2.imread(os.path.join(path, img_path))
        data = img_to_array(img)
        samples = expand_dims(data, 0)
        datagen = ImageDataGenerator(
                                                            rotation_range=15,
                                                            zoom_range=[0.5,1.0],
                                                            fill_mode='nearest',
                                                            horizontal_flip=True,
                                                            vertical_flip=False,
                                                        )
        it = datagen.flow(samples, batch_size=1)
        for i in range(10):
            batch = it.next()
            image = batch[0].astype('uint8')
            cv2.imwrite(os.path.join(SAVE_DIR, CATEGORY, f'image{j}.png'), image)
            j += 1
```
