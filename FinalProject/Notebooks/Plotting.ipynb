```python
%cd /content/drive/MyDrive/CS114.M21.KHCL/DoAn/Model/YOLOv4/darknet
import glob2
import os

files_train = []

image_files = glob2.glob(os.path.join("MyData/train/", "*.jpg"))

files_train += image_files

print(len(files_train))
```

```python
files_test = []

image_files = glob2.glob(os.path.join("MyData/test/", "*.jpg"))

files_test += image_files

print(len(files_test))
```

```python
files_valid = []

image_files = glob2.glob(os.path.join("MyData/valid/", "*.jpg"))

files_valid += image_files

print(len(files_valid))
```

```python
import pandas as pd
df = pd.DataFrame({
    "Size": [len(files_train), len(files_valid), len(files_test)],
    "Type": ["Train", "Valid", 'Test']
})
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
train = len(files_train)
test = len(files_test)
valid = len(files_valid)
labels = ['Train', 'Valid', 'Test']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
bar1 = ax.bar(0, train, 0.35, label='Train')
bar2 = ax.bar(1, valid, 0.35, label='Validation')
bar3 = ax.bar(2, test, 0.35, label='Test')
ax.set_ylabel('Count', fontweight ='bold', fontsize = 15)
ax.set_xlabel('Type', fontweight ='bold', fontsize = 15)


ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.legend()
plt.show()
```

```python

with open("train.txt") as f:
  data_train = f.readlines()
print(len(data_train))

with open("valid.txt") as f:
  data_valid = f.readlines()
print(len(data_valid))

with open("test.txt") as f:
  data_test = f.readlines()
print(len(data_test))
```

```python
c0, c1 = 0, 0
for row in files_train:
    with open('/content/drive/MyDrive/CS114.M21.KHCL/DoAn/Model/YOLOv4/darknet/' + row[:-4] + ".txt") as f:
        data = f.readlines()
    for d in data:
        if d[0] == "0":
            c0 += 1
        elif d[0] == "1":
            c1 += 1
print("TRAIN")
print(f"ChayLa: {c0}")
print(f"DomTrang: {c1}")
```

```python
c2, c3 = 0, 0
for row in data_valid:
    with open('/content/drive/MyDrive/CS114.M21.KHCL/DoAn/Model/YOLOv4/darknet/' + row[:-4] + "txt") as f:
        data = f.readlines()
    for d in data:
        if d[0] == "0":
            c2 += 1
        elif d[0] == "1":
            c3 += 1
print("VALID:")
print(f"ChayLa: {c2}")
print(f"DomTrang: {c3}")
```

```python
c4, c5 = 0, 0
for row in data_test:
    with open('/content/drive/MyDrive/CS114.M21.KHCL/DoAn/Model/YOLOv4/darknet/' + row[:-4] + "txt") as f:
        data = f.readlines()
    for d in data:
        if d[0] == "0":
            c4 += 1
        elif d[0] == "1":
            c5 += 1
print("TEST:")
print(f"ChayLa: {c4}")
print(f"DomTrang: {c5}")
```

```python

import numpy as np
import matplotlib.pyplot as plt
ChayLa = [c0, c2, c4]
DomTrang = [c1, c3, c5]
br1 = np.arange(len(ChayLa))
br2 = [x + 0.25 for x in br1]
plt.bar(br1, ChayLa, color ='r', width = 0.25,
        edgecolor ='grey', label ='ChayLa')
plt.bar(br2, DomTrang, color ='b', width = 0.25,
        edgecolor ='grey', label ='DomTrang')


plt.xlabel('Dataset', fontweight ='bold', fontsize = 15)
plt.ylabel('Count', fontweight ='bold', fontsize = 15)

plt.xticks([r + 0.25 for r in range(len(ChayLa))],
        ['Train', 'Valid', 'Test'])

plt.legend()
plt.show()
```
